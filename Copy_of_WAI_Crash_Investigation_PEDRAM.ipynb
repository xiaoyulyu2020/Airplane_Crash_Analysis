```python
cause_cats={"Weather":["rain","storm","snow","wind","thunderstorm","lightning","fog","icing","weather condition"],
            "Combat":["shot","combat"],
            "Collision":["collide","collision"],
            "technical_failure":["failure","fuel","engine"],
            "Landing":["landing","approach","land"],
            "Take_off":["takeoff","take off","taking off"]
         }
causes=[cause_cat for cause_cat in cause_cats]
threshold=0.95

for cause_cat in cause_cats:
  filter=df["Summary"].str.contains('|'.join(cause_cats[cause_cat]),case=False,regex=True,na=False)
  df[cause_cat] = filter

# df["cause"]=df["Weather"]+df["Combat"]+df["Collision"]+df["technical_failure"]+df["Landing"]+df["Take_off"]
# df[df["cause"]==False].to_csv('remainingdf.csv')
# df.to_csv('remainingdf.csv')
# files.download('remainingdf.csv')

df_saving=df[df["Fatalities_Percent"]<=threshold]
df_fatal=df[df["Fatalities_Percent"]>threshold]

fig, axs = plt.subplots(1,2,figsize=(10, 4))
df_saving[causes].sum().sort_values(ascending=True).plot.barh(ax=axs[0],title=f"Crash Causes - Fatalities less than {threshold*100}%")
df_fatal[causes].sum().sort_values(ascending=True).plot.barh(ax=axs[1],title=f"Crash Causes - Fatalities more than {threshold*100}%")
plt.tight_layout();
```

```python
threshold=0.1
result = pd.Series(index=causes)
for cause_cat in cause_cats:
  df_cause=df[df[cause_cat]==True]
  n_rescue=(df_cause["Fatalities_Percent"]<threshold).sum()
  result[cause_cat]=n_rescue/len(df_cause)

result.sort_values(ascending=False).plot.bar(ylabel="Probability of rescue %",title=f"Probability of rescue\n(Rescue means less than {threshold*100}% fatality)");
```

```python
# in previous section we observed a meaningful rise in airline crashes durring cold war era
# we can go deeper and investigate the crash cause of airlines in this specific period
df_airline=df[df["Operator Cat"]=="Airline"]
result=pd.DataFrame(index=labels,columns=causes)
fig,axs=plt.subplots(figsize=(20,5))

for cause in causes:
  df_airline_cause=df_airline[df_airline[cause]==True]
  df_airline_cause.groupby("Year")["Fatalities"].size().plot(label=cause)
  bin=pd.cut(df_airline_cause.Year, year_bins, labels=labels, right=False)
  case_bin=df_airline_cause.groupby(bin)["Fatalities"].size()#/bin_duration
  result[cause]=case_bin

plt.legend(title='cause')
plt.tight_layout();

fig, axs = plt.subplots(1,2,figsize=(25, 5))
result.div(bin_duration,axis="index").plot(kind="bar",ax=axs[0],title="Crash per Year");
bin_airline=pd.cut(df_airline.Year, year_bins, labels=labels, right=False)
Total_year_bin=df_airline.groupby(bin_airline)["Fatalities"].size()
result_percentage=result.div(Total_year_bin,axis="index").plot(ax=axs[1],title="Share of Crash Cause");
```

```python
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

import folium
from folium.plugins import MarkerCluster,DualMap
from folium import plugins
folium.__version__
```

```python
geolocator = Nominatim(user_agent="geoapiExercises")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1.5)

# according to Nominatim Usage Policy, bulk geocoding is not encouraged,
# therefore we limit number of queries to 100
df.loc[0:100,"geopyLoc"]=df.loc[0:100,"Location"].apply(geocode)
```

```python
df["lat"]=df["geopyLoc"].apply(lambda loc: loc.latitude if isinstance(loc, geopy.location.Location) else None)
df["long"]=df["geopyLoc"].apply(lambda loc: loc.longitude if isinstance(loc, geopy.location.Location) else None)
df_geo=df.dropna(subset=['lat','long'])
```

```python
# map = folium.Map([0, 0], zoom_start=2, control_scale=True, width="%100",height="%50")
map = DualMap([0, 0], zoom_start=2, control_scale=True)
marker_cluster = MarkerCluster().add_to(map)
heat_data=[]
heat_data_fatal=[]

for i,row in df_geo.iterrows():
  popup_txt=f'Date: {row["Date"]}\nFatalities: {row["Fatalities"]}'
  if row["Operator Cat"]=="Military":
    ico=folium.Icon(color="red", icon='fighter-jet', prefix='fa')
  elif row["Operator Cat"]=="Airline":
    ico=folium.Icon(color="blue", icon='plane', prefix='fa')
  else:
    ico=folium.Icon(color="green", icon='question-circle', prefix='fa')

  folium.Marker(location=[row["lat"], row["long"]], popup=popup_txt,tooltip=row["Location"],icon=ico).add_to(marker_cluster)
  heat_data.append((row["lat"], row["long"]))
  heat_data_fatal.append((row["lat"], row["long"],row["Fatalities"]))
  
map.add_child(plugins.MiniMap())
folium.plugins.HeatMap(heat_data).add_to(map.m1)
folium.plugins.HeatMap(heat_data_fatal).add_to(map.m2)
map
```
